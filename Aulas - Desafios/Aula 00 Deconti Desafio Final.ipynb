# **Preparação: Funções**

```
# Gerar grafico de barras
def funcaoGraficoBarras (graficoCor, graficoTitulo, graficoEixoX, graficoEixoY):
  plt.bar(graficoIndex, graficoValues, color=graficoCor)
  plt.title(graficoTitulo, fontweight='bold', fontsize=20)
  plt.xlabel(graficoEixoX, fontweight='bold')
  plt.ylabel(graficoEixoY, fontweight='bold')
  plt.legend('')
  plt.grid(True)
```

```
# Determinar o nome da região com base no código do estado
def funcaoDeterminaRegiao(registro):

    # Define listas das regiões de acordo com o código do estado
    listNorte = [11, 12, 13, 14, 15, 16, 17]
    listNordeste = [21, 22, 23, 24, 25, 26, 27, 28, 29]
    listSudeste = [31, 32, 33, 34, 35]
    listSul = [41, 42, 43]
    listCentroOeste = [51, 52, 53]

    if registro['CO_UF_RESIDENCIA'] in listNorte:
       return 'Norte'

    if registro['CO_UF_RESIDENCIA'] in listNordeste:
       return 'Nordeste'

    if registro['CO_UF_RESIDENCIA'] in listSudeste:
       return 'Sudeste'

    if registro['CO_UF_RESIDENCIA'] in listSul:
       return 'Sul'

    if registro['CO_UF_RESIDENCIA'] in listCentroOeste:
       return 'Centro-Oeste'
```
