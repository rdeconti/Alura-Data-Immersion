**Definindo bibliotecas**

```
# Bilbioteca Matplotlib: criação de gráficos e visualizações de dados em geral
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline

import numpy as np

import pandas as pd

import seaborn as sns
sns.set(color_codes=True)
```

**Obter dados**

```
# Defining source of data (sample provided by Alura)
source = "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

# Reading in the data
dataFrame = pd.read_csv(source)

# Checking number of lines and columns from dataFrame
shape = dataFrame.shape
print ("Número de regristros e número de colunas", shape)
```

**Desafio 01: Obter a proporção dos inscritos por idade**

```
# Getting total number of subscribers
subscribers = len(dataFrame)
print ("Number of subscribers: ", subscribers)

# Compute percentage of subscribers by age and sort ascending by Age
dataSerie = dataFrame["NU_IDADE"].value_counts().sort_index()/subscribers*100

# Compute total number of ages from sata serie
numberOfAges = len(dataSerie)
print ("Number of numberOfAges: ", numberOfAges)
print (" ")

totalPercentage = 0
totalAges = 0

# Print matrix from data serie
for age, quantity in dataSerie.iteritems():
  print(f"Idade {age}  = {quantity:.3f}%")
  totalPercentage = totalPercentage + quantity
  totalAges = totalAges + 1

print (" ")
print ("Checking total of percentage: ", totalPercentage)
print ("Checking total of ages: ", totalAges)
```

**Desafio 02: Descobrir de quais estados são os inscritos com 13 anos**

```
# Performing query from database by State searching by Age equal 13
dataFrame[["SG_UF_RESIDENCIA","NU_IDADE"]].query('NU_IDADE == 13')
```

**Desafio 03: Adicionar título no gráfico**

```
# Generating histogram
dataFrame["NU_IDADE"].value_counts().sort_index()
dataFrame["NU_IDADE"].hist(figsize = (10, 8))

# Setting histogram title
plt.grid(True)
plt.title('Número de inscritos por idade',fontsize = 25)
```

**Desafio 04: Sobreposição dos histogramas de idades de treineiros e não treineiros**

```
#Generating first graph: "NÃO TREINEIRO"
dataFrame.query('IN_TREINEIRO ==0')["NU_IDADE"].hist( bins = 30, figsize = (10, 8),label='Não treineiros')

#Generating second graph: "NÃO TREINEIRO"
dataFrame.query('IN_TREINEIRO ==1')["NU_IDADE"].hist( bins = 30, figsize = (10, 8),label='Treineiros')

#Setting graph parameters
plt.xlabel('Idade',fontsize = 18)
plt.ylabel('Número de inscritos',fontsize = 18)
plt.title('Número de inscritos por idade \n Não Treineiros e Treineiros',fontsize = 25)
plt.legend(('Não treineiros', 'Treineiros'), loc='upper right', shadow = True,fontsize = 20)

#Show graph
plt.show()
```

**Desafio 05: Comparar as distribuições das provas em inglês espanhol**


```
# Compute percentage of subscribers by foreign language
dataSerie = (dataFrame["TP_LINGUA"].value_counts()/dataFrame["TP_LINGUA"].value_counts().sum()) *100
dataSerie.sort_index()
dataSerie.plot.pie(legend= False, title = "Inscritos por língua estrangeira ", autopct='%1.1f%%',labels=['Espanhol','Inglês'],shadow=False )
```

Desafio 06: Comparar as distribuições das provas em inglês espanhol com BOXPLOT

```
fieldList = ["NU_NOTA_LC"]
dataFrame.boxplot(by=['TP_LINGUA'], column = fieldList, grid=True, figsize=(14,12))
plt.title('Comparação as distribuições das provas em inglês espanhol')
plt.xlabel('Língua', fontsize="large")
plt.ylabel('Nota', fontsize='xx-large')
plt.xticks([1,2],['Inglês','Espanhol'])
plt.show()
```
